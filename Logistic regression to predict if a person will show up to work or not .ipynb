```python
#the logistic regression model to predict if a person will be present or absent for work 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import tensorflow as tf 
sns.set()
```

```python
preprocess = pd.read_csv('Absenteeism_preprocessed.csv')
preprocess.head()
```

```python
#create the targets 
#every-value below the median is condiered normal 
#every-value above the median is considered excecess in terms of hours absent 
preprocess['Absenteeism Time in Hours'].median()
```

```python
targets = np.where(preprocess['Absenteeism Time in Hours']>3,1,0)
```

```python
targets
```

```python
preprocess['Absenteeism Excess']=targets
```

```python
preprocess.head()
```

```python
targets.sum()/targets.shape[0]
```

```python
data_with_targets = preprocess.drop(['Absenteeism Time in Hours'],axis=1)
```

```python
data_with_targets.head()
```

```python
#select the inputs for regression 
data_with_targets.shape
```

```python
data_with_targets.iloc[:,0:14]
```

```python
unscaled_inputs = data_with_targets.iloc[:,0:14]
unscaled_inputs.head()
```

```python
#standardise the data 
from sklearn.preprocessing import StandardScaler 
```

```python
absent_scaler = StandardScaler()
```

```python
absent_scaler.fit(unscaled_inputs)
```

```python
scaled_inputs = absent_scaler.transform(unscaled_inputs)
```

```python
scaled_inputs
```

```python
scaled_inputs.shape
```

```python
#train test split 
from sklearn.model_selection import train_test_split 
```

```python
train_test_split(scaled_inputs,targets)
```

```python
x_train,x_test,y_train,y_test = train_test_split(scaled_inputs,targets,train_size = 0.8 ,random_state=52)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#logistic regression with sklearn 
from sklearn import metrics 
from sklearn.linear_model import LogisticRegression 
```

```python
#train the model 
reg = LogisticRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```

```python
#manual check for accuracy 
model_outputs = reg.predict(x_train)
```

```python
model_outputs
```

```python
y_train
```

```python
model_outputs == y_train
```

```python
np.sum(model_outputs == y_train)
```

```python
model_outputs.shape[0]
```

```python
accuracy_check = (437/500)*100
print(round(accuracy_check)," %")
```

```python
#finding the intercepts and coeff 
reg.intercept_
```

```python
reg.coef_
```

```python
unscaled_inputs.columns.values
```

```python
feature_name = unscaled_inputs.columns.values 
```

```python
summary_table = pd.DataFrame(columns=['feature_name'],data = feature_name)
```

```python
summary_table['Coef']=np.transpose(reg.coef_)
```

```python
summary_table
```

```python
summary_table.index = summary_table.index+1
```

```python
summary_table.loc[0] = ['Intercept',reg.intercept_[0]]
```

```python
summary_table=summary_table.sort_index()
summary_table
```

```python
#interpret the coeff 
summary_table['Odds_ratio'] = np.exp(summary_table.Coef)
```

```python
summary_table.sort_values('Odds_ratio',ascending=False)
```

```python
#testing the model 
reg.score(x_test,y_test)
```
